```
import matplotlib.pyplot as plt

import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

img=cv.imread('nature.jpeg')
               
cv2_imshow(img)
```

```
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
#laplacian #takes in source image and data depth 
lap=cv.Laplacian(gray,cv.CV_64F) #transitionong of black to white and vice versa are considered -ve to +ve slope etc
lap=np.uint8(np.absolute(lap))
cv2_imshow(lap)
```

```
#sobel
sobelx=cv.Sobel(gray,cv.CV_64F,1,0) # gradient along x direction
sobely=cv.Sobel(gray,cv.CV_64F,0,1)
cv2_imshow(sobelx)
cv2_imshow(sobely)
```

```
combined_sobel=cv.bitwise_or(sobelx,sobely)
cv2_imshow(combined_sobel)
```

```
canny=cv.Canny(gray,150,175) #canny edge detector#adv algo which uses sobel in 1 of its stages and gives out the output
cv2_imshow(canny)
```
