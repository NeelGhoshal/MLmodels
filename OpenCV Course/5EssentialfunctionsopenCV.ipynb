```
import cv2 as cv
img=cv.imread('nature.jpeg')
from google.colab.patches import cv2_imshow
cv2_imshow(img)
cv.waitKey(0)
```

```
#converting img into grayscale from Bgr type
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
#blurring the img(removes noise)
blur=cv.GaussianBlur(img,(7,7),cv.BORDER_DEFAULT) #args are img,kernel size(always odd num),border func
#to increase blur increase the kernelsize
cv2_imshow(img)
```

```
#edge cascade
canny=cv.Canny(img,125,175)
cv2_imshow(canny) 
```

```
#we can reduce the edges by blurring the img
canny=cv.Canny(blur,125,175)
cv2_imshow(canny) 
```

```
#Dilating the image using structured element
dilated=cv.dilate(canny,(7,7),iterations=3) #structured element,kernel value,iterations
cv2_imshow(dilated) 
```

```
#eroding
eroded=cv.erode(dilated,(7,7),iterations=3) #attempt to get structured element from dilated img
cv2_imshow(eroded)
```

```
#Resize
resized=cv.resize(img,(500,500),interpolation=cv.INTER_LINEAR) #args are img,desired resized dimensions
#INTER_LINEAR AND CUBIC are used when upsizing the img and INTER_AREA used when downsizing

cv2_imshow(resized)
```

```
#cropping
cropped=img[50:200,200:400]
cv2_imshow(cropped)
```
