```
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

img=cv.imread('nature.jpeg')
cv2_imshow(img)
```

```
#A kernel is a window drawn over a specific part of an image(which has a size called as kernel size(i.e. rowas and coloumns of the kernel matrix))
#blur is applied on the middle pixels as a result of the surrounding pixels(eg middlemost part)
#the kernel matrix then slides right and down and does its function over the entire image

```

```
#averaging #takes average of all surrounding pixel intensity and applies it to the current cell
avg=cv.blur(img,(3,3))
cv2_imshow(avg)
```

```
avg=cv.blur(img,(7,7)) #matrix size of kernel changed
cv2_imshow(avg)
```

```
#Gaussianblur avg of products of weights of surrounding pixels are calculated for the current cell
#less blurring 
gaus=cv.GaussianBlur(img,(7,7),0) #args are img,kernel matrix, st. deviation in the x axis
cv2_imshow(gaus)
```

```
#median blur
#median of surrounding blur is found #better to remove noise in the image#not good for high kernel size
median=cv.medianBlur(img,3) #here kernel size is 3 and opencv assumes it to be a 3 by 3 matrix
cv2_imshow(median)
```

```
#bilateral blurring
#used in high level projects
#blurs the image , but retains the edges
bilateral=cv.bilateralFilter(img,5,15,15) #args=img,number,sigma color,sigma space #higher value of sigma space means far away pixels of the image will influence the calculation of the current cell
cv2_imshow(bilateral)
```
