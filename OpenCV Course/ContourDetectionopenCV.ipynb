```
#contours join continous points along the boundaries of an object(can be considered same as edges(not mathametically))
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

img=cv.imread('nature.jpeg')
cv2_imshow(img)
```

```
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
canny=cv.Canny(img,125,175)
cv2_imshow(canny)
```

```
contours,hierarchies=cv.findContours(canny,cv.RETR_TREE,cv.CHAIN_APPROX_NONE) #args for fincountours are structured img and method of finding it
#the above function return a)contours:list of all coordinates of countours in img and b)heirarchies representation of contour
#RETR_LIST will give all contours in a list #RETR_EXTERNAL will give only the external contours #RETR_TREE will retuen the contours in a heirarchial manner
#CHAIN_APPROX_NONE returns the contours(eg.returns all contour points of line, if line is given) #CHAIN_APPROX_SIMPLE compresses the list of contours and return them(eg gives 2 end points if img is line)
print(len(contours))
print("no of contours found")
```

```
contours,hierarchies=cv.findContours(canny,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE) #here it is same
print(len(contours))
print("no of contours found")
```

```
#trying on blur img
blur=cv.GaussianBlur(gray,(5,5),cv.BORDER_DEFAULT)
canny=cv.Canny(blur,125,175)
contours,hierarchies=cv.findContours(canny,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE) #here it is same
print(len(contours))
print("no of contours found")
cv2_imshow(canny)
```

```
#using threshold instead of canny #thresholding attempts to binaries the img to black and white
ret,thresh=cv.threshold(gray,125,255,cv.THRESH_BINARY) 
contours,hierarchies=cv.findContours(canny,cv.RETR_TREE,cv.CHAIN_APPROX_SIMPLE) #here it is same
print(len(contours))
print("no of contours found")
cv2_imshow(thresh)

```

```
blank=np.zeros(img.shape,dtype='uint8')
cv2_imshow(blank) #creating blank img
```

```
#to draw the contours on another img
cv.drawContours(blank,contours,-1,(0,0,255),1) #args are img to draw over,list of contours,indexing of contours to use(-1 is full list),color code,thickness
cv2_imshow(blank)
```
