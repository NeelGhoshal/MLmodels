```
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from google.colab.patches import cv2_imshow

img=cv.imread('nature.jpeg')
cv2_imshow(img)
plt.imshow(img)
plt.show #this will take it in RGB format 
```

```
#color spaces are a space of colors(system of representing array of the img)
#opencv on default uses bgr format for reading imgs
```

```
#BGR to grayscale
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```
#BGR to HSV (hue saturation value)
hsv=cv.cvtColor(img,cv.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

```
#BGR to LAB/l*a*b
lab=cv.cvtColor(img,cv.COLOR_BGR2LAB)
cv2_imshow(lab)
```

```
#BGR to RGB
rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
cv2_imshow(rgb)
plt.imshow(rgb)
plt.show() #here matplotlib deafult is rgb so no inversion
```
