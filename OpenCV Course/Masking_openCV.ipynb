```
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

img=cv.imread('nature.jpeg')
blank=np.ones(img.shape[0:2],dtype='uint8')  #note that the dimensions of the mask must be of the same size of the image 
cv2_imshow(blank)                #for eg, if its grayscale then dimesion should be 1 if bgr then dimension shpuld be 3


cv2_imshow(img)
```

```
mask=cv.circle(blank,(img.shape[1]//2,img.shape[0]//2),100,255,-1)
cv2_imshow(mask)
```

```
masked=cv.bitwise_and(img,img,mask=mask)
cv2_imshow(masked)
```

```
mask=cv.circle(blank,(img.shape[1]//2+45,img.shape[0]//2),100,255,-1)

masked=cv.bitwise_and(img,img,mask=mask)
cv2_imshow(masked)
```
