```
#binarisation of image i.e. take an image an convert the pixels into 0(black) or 255(white)\
```

```
import matplotlib.pyplot as plt

import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

img=cv.imread('nature.jpeg')
               
cv2_imshow(img)
```

```
#eg would be to take an image and take some particular thresholding value and compare each pixel intensity to the thrsholdided values(if equal then 0 and if greater than 255)
```

```
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```
#simple thresholding
thresholding,thresh=cv.threshold(gray,150,255,cv.THRESH_BINARY) #args are img,threshold value,value we want if individual pixel > threshold value(in this casewe want white so 255)
                                            #and thresholding type(here its binary)  
cv2_imshow(thresh)                                      
```

```
#another eg(changed threshold value)
thresholding,thresh=cv.threshold(gray,100,255,cv.THRESH_BINARY) #args are img,threshold value,value we want if individual pixel > threshold value(in this casewe want white so 255)
                                            #and thresholding type(here its binary)  
cv2_imshow(thresh)    
```

```
#another eg
thresholding,thresh=cv.threshold(gray,200,255,cv.THRESH_BINARY) #args are img,threshold value,value we want if individual pixel > threshold value(in this casewe want white so 255)
                                            #and thresholding type(here its binary)  
cv2_imshow(thresh)    
```

```
#create an inverse thresholded img
thresholding,thresh_inv=cv.threshold(gray,150,255,cv.THRESH_BINARY_INV) #args are img,threshold value,value we want if individual pixel > threshold value(in this casewe want white so 255)
                                            #and thresholding type(here its binary)  
cv2_imshow(thresh_inv)    
```

```
#adaptive thresholding
#its better then before as the computer itself finds the best thresholding value(by rg. calculating mean of surrounding pixels)
adaptive_thresh=cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,0)#arg is img,max value,method,type,kernel value,c value(int subtracetd from mean(to fine tune the threshold))
cv2_imshow(adaptive_thresh) 
```

```
#for inv
adaptive_thresh=cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY_INV,11,0)#arg is img,max value,method,type,kernel value,c value(int subtracetd from mean(to fine tune the threshold))
cv2_imshow(adaptive_thresh) 
```

```
#instead of mean 
adaptive_thresh=cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,11,0)#arg is img,max value,method,type,kernel value,c value(int subtracetd from mean(to fine tune the threshold))
cv2_imshow(adaptive_thresh) 
```
