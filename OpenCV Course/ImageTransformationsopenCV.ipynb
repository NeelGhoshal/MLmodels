```
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow

img=cv.imread('nature.jpeg')
cv2_imshow(img)
```

```
#translation
def translate(img,x,y): #x and y are number of pixels we want to shift in x axis and y axis resp
  transMat=np.float32([[1,0,x],[0,1,y]])
  dimensions=(img.shape[1],img.shape[0]) #tuple containg width and height of image
  return cv.warpAffine(img,transMat,dimensions) #-ve values of x implies transforming towards the left
  #-ve values of y means shifting/transforming to up and vice versa

translated=translate(img,100,100) #shift right by 100 pixels and down by 100 pixels
cv2_imshow(translated) 
```

```
#rotation
def rotate(img,angle,rotpoint=None):
  (height,width)=img.shape[0:2]
  if rotpoint is None: #if no value is given when func is called, we will rotate by the center
    rotPoint=(width//2,height//2)
  rotMat=cv.getRotationMatrix2D(rotpoint,angle,1.0)#rotation matrix of the img, with args as rotpoint,angle to be rotated by,scaling
  dimensions=(width,height)
  return cv.warpAffine(img,rotMat,dimensions)
rotated=rotate(img,-45)
cv2_imshow(rotated)     #image is img rotation of translated part, # the upper part to get perfect rotation 
```

```
#resizing
resized=cv.resize(img,(500,500),interpolation=cv.INTER_LINEAR)
cv2_imshow(resized)
```

```
#flipping
flip=cv.flip(img,0) #args are 0,1,-1 for flip code 
cv2_imshow(flip)
```

```
flip=cv.flip(img,1) #args are 0,1,-1 for flip code 
cv2_imshow(flip)
```

```
flip=cv.flip(img,-1) #args are 0,1,-1 for flip code 
cv2_imshow(flip)
```
