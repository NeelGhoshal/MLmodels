**Sigmoid Function**

```python
import math
def sigmoid(x):
  return 1/(1+math.exp(-x))
  
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(-56)
```

**Tanh Function**

```python
def tanh(x):
  return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
  
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
tanh(1)
```

**ReLU**

```python
def ReLU(x):
  return max(0,x)
```

```python
ReLU(-4)
```

```python
ReLU(5)
```

**Leaky ReLU**

```python
def Leaky(x):
  return max(0.1*x,x)
  
```

```python
Leaky(0)
```

```python
Leaky(2)
```

```python
Leaky(-9)
```
