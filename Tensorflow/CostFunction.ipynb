**Mean Absolute Error**

```python
import numpy as np

y_predicted=np.array([1,1,0,0,1])
y_true=np.array([0.3,0.7,1,0,0.5])
```

```python
def mae(y_true,y_predicted):
  totalerror=0
  for yt,yp in zip(y_true,y_predicted):
    totalerror+=abs(yt-yp)
  print(f'total error is {totalerror}')
  mae=totalerror/len(y_true)
  print(f'mae is {mae}')
```

```python
mae(y_true,y_predicted)
```

```python
np.mean(np.abs(y_predicted-y_true)) #numpy operation to find mean absolute error
```

**Binary Cross Entropy/Log Loss**

```python
# note: log 0 dosent exist, so for any possibility of log(0) appearing in a situation, we avert it by replacing the value with a value very close to 0 
```

```python
epsilon=1e-15 #15 0's then 1 (low decimal value)
y_predicted_new=[max(i,epsilon) for i in y_predicted]
y_predicted_new

```

```python
y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
y_predicted_new
```

```python
y_predicted_new=np.array(y_predicted_new)
np.log(y_predicted_new)
```

```python
np.log(y_predicted) #wrong values(log(0))
```

```python
-np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```

```python
def logloss(y_true,y_predicted):
  epsilon=1e-15
  y_predicted_new=[max(i,epsilon) for i in y_predicted]
  y_predicted_new=[min(i,1-epsilon) for i in y_predicted_new]
  y_predicted_new=np.array(y_predicted_new)
  return -np.mean(y_true*np.log(y_predicted_new)+(1-y_true)*np.log(1-y_predicted_new))
```
