```python
from bs4 import BeautifulSoup
import requests

with open('simple.html') as html_file:  #assuming a html file with headings and links and summaries which we want to parse
    soup=BeautifulSoup(html_file,'lxml')
print(soup.prettify())  #.prettify makes the correct nesting indentations(or else soup will be unindeneted and plain)

```

```python
match=soup.title
print(match)
```

```python
match=soup.title.text
print(match)
```

```python
match=soup.div  #gets the first div tag of the file
print(match)
```

```python
match=soup.find('div',class_='footer') #matches a div tag with the attribute of footer
print(match)
```

```python
article=soup.find('div',class_='article') #inspect element in the site and check what you want
print(article)
```

```python
#now we want to print the text at h2 so we can use article directly
headline=article.h2.a.text
print(headline)
```

```python
for article in soup.find_all('div',class_='article'): #for loops thorugh all tags with the parameter(instead of only 1 in find )
  headline=article.h2.a.text
  print(headline)
  summary=article.p.text
  print(summary)
  print()
```
