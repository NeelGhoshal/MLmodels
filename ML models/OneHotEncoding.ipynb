```
import pandas as pd
df=pd.read_csv("houseprices.csv")
df
```

```
dummies=pd.get_dummies(df.town) #creates dummy variables for the specified coloumn
dummies
```

```
merged=pd.concat([df,dummies],axis='columns')   #pandas allows concatenation of 2 arrays of data into 1 dataframe
merged
```

```
final=merged.drop(['town','west'],axis='columns')
final
```

```
from sklearn.linear_model import LinearRegression
model=LinearRegression()
x=final.drop('price',axis='columns') #independent variables
x    
```

```
y=final.price
y
```

```
model.fit(x,y) #training set
```

```
model.predict([[2800,1,0]]) #according to given variables in x
```

```
model.score(x,y) #calculates predictions of all values for input and compares them and finds accuracy of prediction
```

```
from sklearn.preprocessing import LabelEncoder #direct procedure to provide onehotencoding
le=LabelEncoder()
dfle=df
dfle.town=le.fit_transform(dfle.town) #the syntax takes the df columns we provide and turns them transforms it into the encoding part we want
dfle
```

```
x=df[['town','area']].values  #training dataset  #using ,values gives array,if not used it gives dataframe
x
```

```
y=dfle.price
y
```

```
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([('town',OneHotEncoder(),[0])],remainder='passthrough')                                     #creating object of the model  #the categorical features part allows Oth coloumn here to be considered when we use fit func later
```

```
x=ct.fit_transform(x) #town deatures are replaced with the dummy variables and prices are included
x
```

```
x=x[:,1:] #drops 0th coloumn as in the syntax,all rows are selected and 0th coloumn is dropped (we do this to avoid dummy variable trap)
x
```

```
model.fit(x,y)
```

```
model.predict([[1,0,2800]])
```
