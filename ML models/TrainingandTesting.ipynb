```
import pandas as pd
df=pd.read_csv("carprices.csv")
df=df.drop(columns='Unnamed: 0')
df
```

```
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df['Mileage'],df['Sell Price($)'],c='b',marker='+')
```

```
x=df[['Mileage','Age(yrs)']]
y=df[['Sell Price($)']]
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
len(x_train) #16 of 20 are being trained 
```

```
len(x_test)
```

```
x_train #changes after each execution if random state func isnt applied
```

```
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
clf.fit(x_train,y_train)
```

```
clf.predict(x_test) #prediction of y value after putting an element of x value in the linear model created due to 
                   #manipulation by training part of the dataset
 
```

```
y_test
```

```
clf.score(x_test,y_test) #compares predicted and actual values(here 93%)
```
