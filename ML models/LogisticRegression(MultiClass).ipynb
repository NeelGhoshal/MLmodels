```
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits #importing preexisting dataset of handwritten digits
```

```
digits=load_digits()
```

```
dir(digits)
```

```
digits.data[0] #representation of the 1st indexed image in the directory (represented image is in the form of array of values)
```

```
plt.gray()
plt.matshow(digits.images[0])
```

```
for i in range(5):
  plt.matshow(digits.images[i])
```

```
digits.target[0:5]  #this is basically y i.e. the target variable for the given inputs as x
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2) #division into test and train variables(to avoid overfitting and bias towards training data)
```

```
len(x_train)
```

```
len(x_test)
```

```
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```
model.fit(x_train,y_train)
```

```
model.score(x_test,y_test) #using x test y predicted is found which is compared to given y test values
```

```
plt.matshow(digits.images[67]) #picked at random
```

```
digits.target[67]
```

```
model.predict([digits.data[67]])  #giving img as input in form of data to the made model
```

```
model.predict(digits.data[0:5]) 
```

```
y_predicted=model.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predicted) #confusion i.e. wrong values found when comparing y predicted and the actual truth i.e. y test

cm
```

```
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```
