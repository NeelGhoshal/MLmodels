```
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```
dir(iris)
```

```
iris.feature_names
```

```
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```
df['target']=iris.target
df
```

```
iris.target_names
```

```
df[df.target==1].head() #amt of rows for which target variable is 1
```

```
df['flower_name']=df.target.apply(lambda x:iris.target_names[x]) #uses the df target and puts target names value in correlation with it i.e. basically creates a new column from a pre existing column
df.head()
```

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df0=df[df.target==0]
df1=df[df.target==1]
df3=df[df.target==2]
```

```
df0.head()
```

```
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue')
```

```
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='green')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue')
```

```
from sklearn.model_selection import train_test_split
```

```
x=df.drop(['target','flower_name'],axis='columns')
```

```
y=df.target
```

```
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```
len(x_train)
```

```
from sklearn.svm import SVC
model=SVC() #we can change the parameters inside SVC , eg model=SVC(C=10) etc.
model.fit(x_train,y_train)
```

```
model.score(x_test,y_test)
```
