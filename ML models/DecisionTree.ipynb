```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv("salaries.csv")
df=df.drop(columns="Unnamed: 0")
df
```

```
inputs=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```
target
```

```
from sklearn.preprocessing import LabelEncoder #provides encoding of 0,1 and 2 etc on the name values
```

```
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```
inputs['company_n']=le_company.fit_transform(inputs['company']) #make a new column in the df and put values of onehotencoding formed by the given company name values
inputs['job_n']=le_company.fit_transform(inputs['job'])
inputs['degree_n']=le_company.fit_transform(inputs['degree'])
inputs
```

```
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
inputs_n   #so here google is encoded as 2 fb as 1 and abc pharma as 0 for the company list(similarly for the other lists)
```

```
from sklearn import tree
```

```
model=tree.DecisionTreeClassifier()
```

```
model.fit(inputs_n,target) #here the criteria 'gini' can be considered as entropy 
```

```
model.score(inputs_n,target)
```

```
model.predict([[2,2,1]]) #works in google,sales exec,masters
```
