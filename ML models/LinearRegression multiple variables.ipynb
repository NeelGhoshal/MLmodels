```
import numpy as np
import pandas as pd
from sklearn import linear_model
df=pd.read_csv("houseprices.csv")
df
```

```
import math
median_bedrooms=math.floor(df.bedrooms.median()) #to get absolute integer value not whole number
median_bedrooms
```

```
df.bedrooms=df.bedrooms.fillna(median_bedrooms)
df
```

```
reg = linear_model.LinearRegression() #make linear reg class object
reg.fit(df[['area','bedrooms','age']],df.price )#train the model
```

```
reg.coef_#m1 m2 and m3
```

```
reg.intercept_
```

```
reg.predict([[3000,3,40]])
```
