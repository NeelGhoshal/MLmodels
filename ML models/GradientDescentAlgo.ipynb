```
import numpy as np
def gradient_descent(x,y):
 m_curr=b_curr=0 #initial values of m and b of line
 iterations=1000 # amt of steps we take in the graph
 n=len(x)
 learning_rate=0.001 #starting randomly given value for this
 for i in range(iterations):
   y_predicted=m_curr*x+b_curr
   cost=(1/n)*sum([val**2 for val in (y-y_predicted) ]) #cost function using list operation on given y value list and y predicted by us
   md=-(2/n)*sum(x*(y-y_predicted))  #derivative wrt m
   bd=-(2/n)*sum(y-y_predicted)
   m_curr=m_curr-learning_rate*md
   b_curr=b_curr-learning_rate*bd
   print("m{},b{},iteration{}".format(m_curr,b_curr,cost,i)) #to know what happens in each iteration
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13]) #we want to find the best fit line for the given values of x and y

gradient_descent(x,y)                                                                                     
```
